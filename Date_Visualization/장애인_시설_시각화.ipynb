###구별 장애인 수 지도에 시각화

```python
!pip install folium
```

```python
import folium
import pandas as pd
import json
import re

geo_json='https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

csv_file_path = "/content/gdrive/MyDrive/문화데이터/장애인_문화스포츠_인프라_모음 (1).csv"
df = pd.read_csv(csv_file_path)

df2=df[['자치구','장애인_합계']]
df2.columns=['name','values']

m=folium.Map(
    location=[37.559819, 126.963895],
    tiles='Stamen Terrain'
)

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=df2,
    columns=['name','values'],
    key_on='properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2
).add_to(m)
m.save('/content/gdrive/MyDrive/문화데이터/map.html')
```

###지하철 엘리베이터, 휠체어 리프트, 급속충전소 누적 막대차트


```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Bar3D

# CSV 파일 경로
csv_file_path = "/content/gdrive/MyDrive/문화데이터/장애인_문화스포츠_인프라_모음 (1).csv"

# CSV 파일 로드
df = pd.read_csv(csv_file_path)
# 원하는 HTML 파일의 경로와 파일명 설정
html_file_path = os.path.join("/content/gdrive/MyDrive/문화데이터", "stacked_bar.html")

# 행 이름 추출
row_names = df["자치구"].tolist()
# x, y, z 축 데이터 추출
y_data1 = df["지하철_엘리베이터_수"].tolist()
y_data2 = df["전동보장구_급속충전소"].tolist()
y_data3 = df["지하철_휠체어리프트_수"].tolist()



# 막대 그래프 생성
bar = (
    Bar()
    .add_xaxis(row_names)
    .add_yaxis("지하철_엘리베이터_수", y_data1, stack="stack1")
    .add_yaxis("전동보장구_급속충전소", y_data2, stack="stack1")
    .add_yaxis("지하철_휠체어리프트_수", y_data3, stack="stack1")
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True))  # 막대 위에 값 표시
    .set_global_opts(title_opts=opts.TitleOpts(title="누적 막대 그래프"))
)

# 그래프 렌더링
bar.render(html_file_path)
```
